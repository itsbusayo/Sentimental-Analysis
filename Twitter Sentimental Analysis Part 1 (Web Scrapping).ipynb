```python
import snscrape.modules.twitter as sntwitter
import pandas as pd

# Creating list to append tweet data to
tweets_list1= []

# Using TwitterSearchScraper to scrape data and append tweets to list
for i,tweet in enumerate(sntwitter.TwitterSearchScraper('Muhammadu Buhari since:2021-10-01 until:2022-01-31').get_items()):
    if i>50000:
        break
    tweets_list1.append([tweet.date, tweet.id, tweet.content, tweet.user.username])
    
# Creating a dataframe from the tweets list above
tweets_df1= pd.DataFrame(tweets_list1, columns=['Datetime', 'Tweet Id', 'Text', 'Username'])
```

```python
tweets_df1
```

```python
# Creating the second list to append tweet data to
tweets_list2 = []

# Using TwitterSearchScraper to scrape data and append tweets to the second list
for i,tweet in enumerate(sntwitter.TwitterSearchScraper('Atiku Abubakar since:2021-10-01 until:2022-01-31').get_items()):
    if i>50000:
        break
    tweets_list2.append([tweet.date, tweet.id, tweet.content, tweet.user.username])
    
# Creating a dataframe from the tweets list above
tweets_df2 = pd.DataFrame(tweets_list2, columns=['Datetime', 'Tweet Id', 'Text', 'Username'])
```

```python
tweets_df2
```

```python
# Export both dataframes into a CSV
tweets_df1.to_csv('Muhammed-Buhari-query-tweets.csv', sep=',', index=False)
tweets_df2.to_csv('Atiku Abubakar-query-tweets.csv', sep=',', index=False)
```

```python
Tweets_Merged = pd.concat([tweets_df1,tweets_df2])
```
